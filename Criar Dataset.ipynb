### Cria uma schema no SQLite

```python
import sqlite3
conn = sqlite3.connect('social.db')

cursor = conn.cursor()

# criando a tabela (schema)
cursor.execute("drop  table social");
cursor.execute("CREATE TABLE social ( Id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT, user INT NOT NULL,slot INT NOT NULL, dayWeek INT NOT NULL, weekend Boolean NOT NULL, day INT NOT NULL, month INT NOT NULL, conversation Boolean NOT NULL, line INT)")
conn.commit()
conn.close()
```

### Define os metodos para criar os slots de conversação

```python
from datetime import datetime
import pytz

def createAllSlots(dataset , dayWeek , weekend , day , month):
    for slot in range(0,24*4):
        row={'slot':slot,'dayWeek':dayWeek,'weekend':weekend,'day':day,'month':month,'conversation':False,'line':0}
        dataset.append(row)
        
    

def checkSlot(dataset,slot,day,month,line):
    for row in dataset:
        if row['slot'] == slot and row['day'] == day and row['month'] == month:
            row['conversation']=True
            row['line']=line
            break 

```

### Ler os dados do arquivo!

```python
def createDataSet(file):
    f = open(file, "r")
    f.readline()
    dataset=[]
    days=[]
    tz = pytz.timezone('US/Eastern')
    line=0
    for l in f:
        line+=1
        tempo=l.replace("\n","").split(",")
        #verifica se o registro possui dois valores: inicio e fim
        if len(tempo)!=2:
            continue
        time1=datetime.fromtimestamp(int(tempo[0]),tz=tz)
        time2=datetime.fromtimestamp(int(tempo[1]),tz=tz)
        
        
        t1Hora=int(time1.strftime('%H'))
        t1Min=int(time1.strftime('%M'))
        
        t2Hora=int(time2.strftime('%H'))
        t2Min=int(time2.strftime('%M'))
        
        day=int(time2.strftime('%d'))
        month=int(time2.strftime('%m'))
  
        slotIni=((t1Hora*4)+int(t1Min/15))
        slotFim=((t2Hora*4)+int(t2Min/15))
        
        
        dayWeek=int(time1.strftime('%w'))
        weekend=(dayWeek==0 or dayWeek==6)
        strDate= str(month)+""+str(day) 
        
        if strDate not in days:
            createAllSlots(dataset,dayWeek,weekend,day,month)
            days.append(strDate)
            
            
        for slot in range(slotIni,slotFim+1):
            checkSlot(dataset,slot,day,month,line)
         
    return dataset

```

### Insere os dados de conversação de todos os usuários em um banco de dados 

```python
#{'slot': 1, 'dayWeek': 3, 'weekend': False, 'day': 27, 'month': 3, 'conversation': False}

def inserir(user,slot,dayWeek,weekend,day,month,conversation,line):
    dados=[(user,slot,dayWeek,weekend,day,month,conversation,line)]
 #   print(dados)
    if(1==1):
        conn = sqlite3.connect('social.db')
        cursor = conn.cursor()
        cursor.executemany("INSERT INTO social (user,slot,dayWeek,weekend,day,month,conversation,line) VALUES (?,?,?,?,?,?,?,?)", dados)
        conn.commit()
        conn.close()
        return True
    return False

```

### Exportar para ARFF

```python
def criarArff(file,dados):
    f=open(file","w")
    f.write("@RELATION mente \n")
    f.write("\n")
    f.write("@ATTRIBUTE slot NUMERIC \n")
    f.write("@ATTRIBUTE dayWeek NUMERIC \n")
    f.write("@ATTRIBUTE weekend {True, False} \n")
    f.write("@ATTRIBUTE conversation {True, False} \n")
    f.write("\n")
    f.write("@DATA \n")

    for line in dados:
        string=str(line['slot'])+","+str(line['dayWeek'])+","+str(line['weekend'])+","+str(line['conversation'])
        f.write(string+'\n')
        print(string)

    f.close()
    
```

```python
criarArff("dataset.arff",dados)
```

### Exportar para CSV

```python
def criarCSV(file,dados):
    f=open(file,"w")
    f.write("slot,dayWeek,weekend,conversation\n")

    for line in dados:
        weekend=0
        conversation=0
        if line['weekend']:
             weekend=1
        if line['conversation']:
            conversation=1

        string=str(line['slot'])+","+str(line['dayWeek'])+","+str(weekend)+","+str(conversation)
        f.write(string+'\n')
        print(string)

    f.close()
```

```python
criarCSV("dataset.csv",dados)
```
